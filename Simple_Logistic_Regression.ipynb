```python
x = df1.iloc[:,:-1]
x
y = df1.iloc[:,-1]
y
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
col = x_train.columns
col
```

```python
from sklearn.preprocessing import StandardScaler


Scale = StandardScaler()
x_train= pd.DataFrame(Scale.fit_transform(x_train),columns=col)
x_train
```

```python
x_test = pd.DataFrame(Scale.fit_transform(x_test),columns=col)
x_test
```

```python
from sklearn.linear_model import LogisticRegression

Model = LogisticRegression()
Model.fit(x_train,y_train)
```

```python
y_pred = Model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix

accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_train,Model.predict(x_train))
```

```python
sns.heatmap(confusion_matrix(y_test,y_pred),annot=True,cmap='Greens',)
```

```python
from sklearn.metrics import roc_curve, roc_auc_score,classification_report
import matplotlib.pyplot as plt
fpr, tpr, thresholds = roc_curve(y_test,y_pred)

plt.figure(figsize=(8,6))
plt.plot(fpr, tpr, label="Roc_Curve".format(roc_auc_score(y_test,y_pred)))
plt.plot([0,1], [0,1])
plt.show()
```

```python
classification_report(y_test,y_pred)
```
